```python
import random
import time

# Word list
words = ["neon", "futuristic", "python", "cyber", "galaxy", "matrix", "streamlit"]
word = random.choice(words).upper()

# Hangman stages with emojis
hangman_stages = [
    "\n🔥🚀🔥 Futuristic Hangman 🔥🚀🔥\n",
    "😃 \n",
    "😐 \n O \n",
    "😕 \n O \n | \n",
    "😟 \n O \n/| \n",
    "😨 \n O \n/|\\ \n",
    "💀 \n O \n/|\\ \n/  \n",
    "☠️ \n O \n/|\\ \n/ \\ \n"
]

attempts = 7
guessed_letters = []
display_word = ["_" for _ in word]

print(hangman_stages[0])
time.sleep(1)

while attempts > 0 and "_" in display_word:
    print("\033[92mWord: " + " ".join(display_word))  # Green text
    print("\033[93mAttempts left: " + str(attempts))  # Yellow text
    print("\033[95mGuessed letters: " + (", ".join(guessed_letters) if guessed_letters else "None"))  # Magenta text
    print("\033[91m" + hangman_stages[7 - attempts])  # Red text

    guess = input("\033[96m\nGuess a letter: ").strip().upper()  # Cyan text

    if not guess or len(guess) != 1 or not guess.isalpha():
        print("\033[91mInvalid input! Please enter a single letter.")
        continue

    if guess in guessed_letters:
        print("\033[91mYou already guessed that letter!")
        continue

    guessed_letters.append(guess)

    if guess in word:
        print("\033[92mCorrect! 🎉")  # Green text
        for i in range(len(word)):
            if word[i] == guess:
                display_word[i] = guess
    else:
        print("\033[91mWrong! ❌")  # Red text
        attempts -= 1

if "_" not in display_word:
    print("\033[92m\n🎉 YOU WIN! The word was: " + word)  # Green text
else:
    print("\033[91m\n💀 GAME OVER! The word was: " + word)  # Red text
```
